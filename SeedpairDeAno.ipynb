```python
import networkx as nx
import math
import numpy as np
import json
```

```python
from google.colab import files #this is used to upload the local files into collab
g1 = files.upload()
g2 = files.upload()
snp = files.upload()
```

```python
def degree(unpaired_node_list,Graph):
  degree={}
  for i in unpaired_node_list:
    degree[i]=Graph.degree(i)
  return degree    
```

```python
g1 = nx.read_edgelist("seed_G1.edgelist", nodetype=int) #g1
g2 = nx.read_edgelist("seed_G2.edgelist", nodetype=int) #g2
sp = open('seed_node_pairs.json')
ini_snp = json.load(sp) #old_g1_pair_dict/nodepairs
#ini_snp = {}
#with open("seed_node_pairs.txt",'r') as f:  
 # node_pairs=f.readlines()
## pair=pair.strip('\n').split(' ')
  #ini_snp[int(pair[0])]=int(pair[1])

ini_g1_sp = [] #old_g1_pair
ini_g2_sp = [] #old_g2_pair
snp = {}
for key, value in ini_snp.items():
  ini_g1_sp.append(int(key))
  ini_g2_sp.append(int(value))
  snp[int(key)] = int(value)
g1_sp = ini_g1_sp.copy()
g2_sp = ini_g2_sp.copy()

#calculating unpaired nodes
unpaired_g1 = []
unpaired_g2 = set([])
for i in g1.nodes:
  if i not in ini_g1_sp:
    unpaired_g1.append(i)
for j in g2.nodes:
  if j not in ini_g2_sp:
    unpaired_g2.add(j)
#calculating degree of unpaired nodes    
unpaired_g1_degree = degree(unpaired_g1,g1)    
unpaired_g2_degree = degree(unpaired_g2,g2)    
```

```python
#check in pairs
for node in unpaired_g1:
  nodepair = set([])
  for neighbor in g1.neighbors(node):
    if neighbor in g1_sp:
      nodepair.add(neighbor)
  g2_node_score = {}
  for j in unpaired_g2:
    other_nodepair = set([])
    for i in nodepair:
      other_nodepair.add(snp[i])
    count = 0
    for x in other_nodepair:
      if x in g2.neighbors(j):
        count +=1     

    score = count/((math.sqrt(unpaired_g1_degree[node]))*(math.sqrt(unpaired_g2_degree[j])))
    g2_node_score[j] = score

  scores = set(g2_node_score.values())
  max1 = max(scores) #max score
  for k, v in g2_node_score.items():
    if v == max1:
      g2_max_score = k
  std = np.std(list(scores))
  scores.remove(max1) 
  max2 = max(scores) 
  if std == 0:
    ecce = 0
  else:
    ecce = (max1 - max2)/std    #ecce score
  if ecce > 0.5:
    g1_sp.append(node)
    g2_sp.append(g2_max_score)
    snp[node]=g2_max_score
    unpaired_g1.remove(node)
    unpaired_g2.remove(g2_max_score)
                       
with open('Indrojuproject1output.txt', 'w') as f:
  for i in g1_sp:
    #z = str(i)+ ' ' + str((snp[i])+'\n']
    z = str(i)+ ' ' + str(snp[i])+'\n'
    f.write(z)
f = open("Indrojuproject1output.txt", "r")
files.download('Indrojuproject1output.txt')

```
